```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
```

```python
df = pd.read_csv('postpro_anal.csv')
df['thr'] = df['thr'].round(2)
df
```

```python
plt.figure()
sns.lineplot(data=df, x='thr', y='max')
plt.ylim(0, 6000)
plt.xlabel('confidence threshold')
plt.ylabel('max volume')
plt.show()
```

```python
plt.figure()
sns.lineplot(data=df, x='thr', y='count', hue='minvol')
plt.ylim(0, 300)
plt.xlabel('confidence threshold')
plt.ylabel('number of components')
plt.show()
```

```python
plt.figure()
sns.lineplot(data=df, x='minvol', y='count', hue='thr')
plt.ylim(0, 300)
plt.xlabel('minimum volume threshold')
plt.ylabel('number of components')
plt.show()
```

Conclusion:

The number of components decreases considerably when minimum_volume_threshold is increased. There are a lot of small components, specially in the volume range [0,40].

```python
plt.figure()
sns.lineplot(data=df, x='thr', y='mean', hue='minvol')
plt.ylim(0, 1000)
plt.xlabel('confidence threshold')
plt.ylabel('mean volume')
plt.show()
```

```python
plt.figure()
sns.lineplot(data=df, x='minvol', y='mean', hue='thr')
plt.plot(range(0, 160), range(0,160), '--', color='red')
plt.ylim(0, 1000)
plt.xlabel('minimum volume threshold')
plt.ylabel('mean volume')
plt.show()
```
